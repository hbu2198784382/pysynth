```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import scipy.signal
```

```python
t = 2
sr = 44100
freq1 = 440*(2**(3/12)) # C4 
freq2 = 440*(2**(7/12)) # E4
freq3 = 440*(2**(10/12)) # G4
freq4 = freq1*2
    
n = int(t * sr)
tn = np.linspace(0,t,n,endpoint=False)
y = np.sign(np.sin(2*np.pi*freq1*tn))/2 + np.sign(np.sin(2*np.pi*freq2*tn))/3
nf = len(y)
f = np.abs(np.fft.rfft(y)) / np.sqrt(n) / 2.0
freqs = np.fft.rfftfreq(n,1.0/sr)


plt.figure(num=1,figsize=(10,5))
plt.plot(tn[:800],y[:800])

plt.figure(num=2,figsize=(10,5))
plt.plot(freqs[:3000],f[:3000])

plt.show()
ipd.display(ipd.Audio(y, rate=sr))
```
